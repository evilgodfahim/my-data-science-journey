```python
%matplotlib inline
import matplotlib.pyplot as plt
x=[3,6,3,1,7,3]
y=[18,27,44,73,23,30]
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title="Simple plot",
      xlabel="X axis",
      ylabel="Y axis")
fig.savefig("image/Sampleplot.png")
```

```python
dic={"Science book":5,"Fiction books":16,"Comic books":15}
```

```python
fig,ax=plt.subplots()
ax.bar(dic.keys(),dic.values());
ax.set(title="Daily Sells",xlabel="Books Catagory",ylabel="Number of Sells");
```

```python
fig,ax=plt.subplots()
ax.barh(list(dic.keys()),list(dic.values()));
```

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(20,20))
ax1.plot(x,np.tan(x));
ax2.plot(x,x**2);
ax3.plot(x,np.sin(x));
ax4.plot(x,np.exp(x))
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(20,10))
ax1.plot(x,np.cos(x))
ax2.bar(x,np.tan(x))
ax3.scatter(x,np.exp(x))
ax4.hist(x)
```
