```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
carsales=pd.read_csv("car sales.csv")
```

```python
carsales
```

```python
carsales["Price"]=carsales["Price"].str.replace("[/$,/,/.]","")
```

```python
carsales
```

```python
carsales["Price"]=carsales["Price"].str[:-2]
```

```python
carsales
```

```python
carsales["Date Range"]=pd.date_range("1/1/2020",periods=len(carsales))
```

```python
carsales
```

```python
carsales["Total sales"]=carsales["Price"].astype(int).cumsum()
```

```python
 carsales
```

```python
%matplotlib inline
carsales.plot("Date Range","Total sales");
```

```python
fig,ax=plt.subplots()
ax.plot(carsales["Date Range"],carsales["Total sales"]);
ax.set(title="Simple plot",xlabel="Date",ylabel="Sales")
```

```python
heart=pd.read_csv("heart.csv")
```

```python
heart
```

```python
heart["age"].plot(kind="hist");
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(heart["age"],heart["sex"])
```

```python
heart.plot.hist(figsize=(10,50),subplots=True);
```

```python
over50=heart[heart["age"]>50]
```

```python
len(over50)
```

```python
over50.reset_index(drop=True)
```

```python
over50.plot(kind="scatter",x="age",y="chol",c="Blue",figsize=(10,5));
```

```python
plt.style.use("seaborn-whitegrid")
fig,(ax1,ax2)=plt.subplots(nrows=2
                           ,ncols=1,
                           figsize=(10,10),
                           sharex=True)

splot=ax1.scatter(x=over50["age"],
                  y=over50["chol"],
                  c=over50["thal"],
                  cmap='winter');

ax1.set(xlim=(45,100),
        title="Age vs Cholesterol level",
        ylabel="Cholesterol");

ax1.legend(*splot.legend_elements(),
           title="Thal");

ax1.axhline(over50["chol"].mean()
          ,linestyle="--",
            c="red");

splot2=ax2.scatter(over50["age"],
                   over50["thalach"],
                   c=over50["age"],
                   cmap="winter");

ax2.set(ylim=(60,200),
        title="Age vs Heart rate",
        xlabel="Age",
        ylabel="Heart Rate");

ax2.legend(*splot2.legend_elements(),
           title="Heart Rate");

ax2.axhline(over50["thalach"].mean(),
            linestyle="--",
            c="Green");
fig.suptitle(t="Heart diesease analysis",fontsize=20,c="blue",fontweight="bold");
```
